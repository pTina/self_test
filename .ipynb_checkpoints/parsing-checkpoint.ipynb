```python

f = open('test01.html','r',encoding='UTF-8')
while True:
    line = f.readline()
    s = "'"+line+"'"+"+"
    if not line:
        break;
    print(s)
```

```python
# open과 close를 자동으로 해주는 것 => with
# width문을 빠져나올 때 자동으로 close를 호출한다.

# 'test01.html'파일을 읽기모드로 file이라는 이름으로 열거다.
with open('test01.html','r',encoding='UTF-8') as file: 
    # file에 라인이 있는동안
    for line in file: 
        print("'"+line.strip('\n')+"'+") # 출력
        
```

```python

array = []
count = 0

with open('test01.html','r',encoding='UTF-8') as file: 
    for line in file:
        array.append(line.strip('\n'))

for i in array:
    print("'"+i+"'+")
    count+=1
    if len(array) == count:
        print("'"+i+"';")
```
